```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv("time_serise.csv")
it=pd.read_csv("ita.csv")
i_d=it['currently_positive']
```

```python
df=df.loc[df['Country/Region']=='Mainland China']
ch=df.drop(columns=['Province/State','Country/Region','Lat','Long'])
```

```python
y=ch.sum().values
x=ch.columns
d = {'date': x, 'infected': y}
d=pd.DataFrame(data=d)
# d=pd.DataFrame([[y,x]],columns=['date','infected'])
d
```

```python
fig, ax = plt.subplots(figsize=(15,8))
g=sns.lineplot(x='date',y='infected',sort= False,data=d)
g.set_xticklabels(labels=x,rotation=90)
```

```python
plt.figure(figsize=(12, 8))

plt.plot(d['infected'],label="China")
plt.plot(i_d,label="Italy")
plt.xticks(rotation=90)
plt.legend(bbox_to_anchor=(1.01, 1), loc='upper left', borderaxespad=0.)
```
